Задача 1. Дан набор из p матриц размерностью(n, n) и p векторов размерностью (n, 1), найти сумму произведений матриц на векторы. Написать тесты для кода

```python
import numpy as np

def sum_prod(X, V):
    if len(X) != len(V):
        raise ValueError("Количество матриц и векторов должно совпадать")

    result = np.zeros_like(V[0])

    for i in range(len(X)):
        result = result + X[i].dot(V[i])

    return result
    
# Тест 1
X_test = [
    np.array([[1,2],[3,4]]),
    np.array([[2,0],[1,2]])
]
V_test = [
    np.array([[1],[2]]),
    np.array([[3],[1]])
]

res = sum_prod(X_test, V_test)
print(res)
assert (res == np.array([[11],[16]])).all()


# Тест 2: единичные матрицы
X_test2 = [np.eye(2), np.eye(2)]
V_test2 = [np.array([[5],[6]]), np.array([[1],[2]])]

res2 = sum_prod(X_test2, V_test2)
print(res2)
assert (res2 == np.array([[6],[8]])).all()


# Тест 3: ошибка размерности списков
try:
    sum_prod([np.eye(2)], [np.eye(2), np.eye(2)])
    assert False
except ValueError:
    assert True
```

Задача 2. Дана матрица M, напишите функцию, которая бинаризует матрицу по некоторому threshold (то есть, все значения большие threshold становятся равными 1, иначе 0). Напишите тесты для кода

```python
import numpy as np

def binarize(M, threshold=0.5):
    A = np.asarray(M)
    return (A > threshold).astype(np.int32)
# 1 тест 
M1 = np.array([[0.1, 0.5, 0.9],
               [0.6, 0.2, 0.49]])
exp1 = np.array([[0, 0, 1],
                 [1, 0, 0]])
assert np.array_equal(binarize(M1, 0.5), exp1)

# 2 Кастомный порог
M2 = np.array([[2.0, 3.0], [4.0, 1.0]])
exp2 = np.array([[0, 0], [1, 0]])   # порог 3.5
assert np.array_equal(binarize(M2, 3.5), exp2)

# 3) 0
M3 = np.array([0.3, 0.7, 0.7, 0.2])
exp3 = np.array([0, 0, 0, 0])
assert np.array_equal(binarize(M3, 0.7), exp3)

# 4) Вход — список списков
M4 = [[-1, 0, 1], [2, -2, 0.4]]
exp4 = np.array([[0, 0, 1], [1, 0, 0]])
assert np.array_equal(binarize(M4, 0.5), exp4)

# 5) Исходная матрица не модифицируется (функция возвращает копию)
M5 = np.array([0.2, 0.9])
_ = binarize(M5, 0.5)
assert np.array_equal(M5, np.array([0.2, 0.9]))
```

Задача 3. Напишите функцию, которая возвращает уникальные элементы из каждой строки матрицы. Напишите такую же функцию, но для столбцов. Напишите тесты для кода

```python
import numpy as np

def unique_rows(mat):
    arr = np.asarray(mat)
    result = []
    for row in arr:
        result.append(np.unique(row))
    return result


def unique_columns(mat):
    arr = np.asarray(mat)
    result = []
    for col in arr.T:
        result.append(np.unique(col))
    return result
# Базовая матрица
M = np.array([
    [1, 2, 2, 3],
    [4, 4, 4, 4],
    [5, 6, 5, 6]
])

# Тест unique_rows
rows_expected = [np.array([1, 2, 3]),
                 np.array([4]),
                 np.array([5, 6])]

rows_res = unique_rows(M)
for res, exp in zip(rows_res, rows_expected):
    assert np.array_equal(res, exp)

# Тест unique_columns
cols_expected = [np.array([1, 4, 5]),
                 np.array([2, 4, 6]),
                 np.array([2, 4, 5]),
                 np.array([3, 4, 6])]

cols_res = unique_columns(M)
for res, exp in zip(cols_res, cols_expected):
    assert np.array_equal(res, exp)

# Тест на список (не только np.array)
M2 = [[7,7],[8,9]]
assert [list(x) for x in unique_rows(M2)] == [[7],[8,9]]
assert [list(x) for x in unique_columns(M2)] == [[7,8],[7,9]]
```

```python
import numpy as np
import matplotlib.pyplot as plt

def analyze_normal_matrix(m, n):
    # создаём матрицу из нормального распределения
    A = np.random.randn(m, n)

    # считаем мат. ожидание и дисперсию по строкам и столбцам
    mean_rows = np.mean(A, axis=1)
    var_rows = np.var(A, axis=1)

    mean_cols = np.mean(A, axis=0)
    var_cols = np.var(A, axis=0)

    print("Матожидание по строкам:", mean_rows)
    print("Дисперсия по строкам:", var_rows)
    print("Матожидание по столбцам:", mean_cols)
    print("Дисперсия по столбцам:", var_cols)

    # гистограммы для строк
    for i in range(m):
        plt.figure()
        plt.title(f"Гистограмма для строки {i+1}")
        plt.hist(A[i, :], bins=10, color='skyblue', edgecolor='black')
        plt.xlabel("Значение")
        plt.ylabel("Частота")
        plt.show()

    # гистограммы для столбцов
    for j in range(n):
        plt.figure()
        plt.title(f"Гистограмма для столбца {j+1}")
        plt.hist(A[:, j], bins=10, color='lightgreen', edgecolor='black')
        plt.xlabel("Значение")
        plt.ylabel("Частота")
        plt.show()

    return A, mean_rows, var_rows, mean_cols, var_cols
```

```python
import numpy as np

def chess(m, n, a, b):
    # создаём пустую матрицу
    mat = np.zeros((m, n), dtype=int)

    for i in range(m):
        for j in range(n):
            # если сумма индексов чётная — ставим a, иначе b
            if (i + j) % 2 == 0:
                mat[i, j] = a
            else:
                mat[i, j] = b

    return mat
# тест 1 — маленькая матрица
print(chess(2, 2, 1, 0))

# тест 2 — прямоугольная матрица
print(chess(3, 4, 5, 9))

# тест 3 — другой набор значений
print(chess(3, 3, 7, -1))
```

```python
import numpy as np

def draw_rectangle(a, b, m, n, rectangle_color, background_color):
    # создаём фон
    img = np.zeros((m, n, 3), dtype=int)
    img[:, :] = background_color

    # центр изображения
    cx, cy = m // 2, n // 2

    # границы прямоугольника
    top = cx - a // 2
    bottom = cx + a // 2
    left = cy - b // 2
    right = cy + b // 2

    # закрашиваем прямоугольник
    img[top:bottom, left:right] = rectangle_color

    return img


def draw_ellipse(a, b, m, n, ellipse_color, background_color):
    img = np.zeros((m, n, 3), dtype=int)
    img[:, :] = background_color

    cx, cy = m // 2, n // 2  # центр

    for i in range(m):
        for j in range(n):
            # формула эллипса
            if ((i - cx) ** 2) / (a ** 2) + ((j - cy) ** 2) / (b ** 2) <= 1:
                img[i, j] = ellipse_color

    return img
    
# тест 1 — прямоугольник по центру
img = draw_rectangle(4, 6, 10, 10, [255,0,0], [0,0,0])
print(np.sum(img == [255,0,0]))   # должно быть > 0


# тест 2 — эллипс
img2 = draw_ellipse(3, 5, 20, 20, [0,255,0], [0,0,0])
print(np.sum(img2 == [0,255,0]))  # тоже > 0


# тест 3 — размеры матрицы
assert draw_rectangle(4,4,20,30,[1,2,3],[0,0,0]).shape == (20,30,3)
assert draw_ellipse(5,5,15,25,[1,2,3],[0,0,0]).shape == (15,25,3)
```

```python
from math import sqrt
from typing import List, Tuple, Dict

def series_stats(x: List[float], p: int) -> Dict[str, object]:
    n = len(x)
    if n == 0:
        return {
            "mean": 0.0,
            "var": 0.0,
            "std": 0.0,
            "local_max_idx": [],
            "local_min_idx": [],
            "moving_avg": []
        }

    # мат. ожидание
    mean = sum(x) / n

    # дисперсия (генеральная)
    var = sum((xi - mean) ** 2 for xi in x) / n
    std = sqrt(var)

    # локальные экстремумы (строгие)
    max_idx = []
    min_idx = []
    for i in range(1, n - 1):
        if x[i] > x[i - 1] and x[i] > x[i + 1]:
            max_idx.append(i)
        if x[i] < x[i - 1] and x[i] < x[i + 1]:
            min_idx.append(i)

    # скользящее среднее (окно p, шаг 1)
    mov = []
    if p > 0 and p <= n:
        window_sum = sum(x[:p])
        mov.append(window_sum / p)
        for i in range(p, n):
            window_sum += x[i] - x[i - p]
            mov.append(window_sum / p)
    # если p некорректен — вернётся []

    return {
        "mean": mean,
        "var": var,
        "std": std,
        "local_max_idx": max_idx,
        "local_min_idx": min_idx,
        "moving_avg": mov
    }
```

```python
def one_hot_encode(labels):
    num_classes = max(labels) + 1
    
    # создаём список списков, инициализированных нулями
    result = []
    for lab in labels:
        row = [0] * num_classes
        row[lab] = 1
        result.append(row)
        
    return result
# тесты
assert one_hot_encode([1]) == [[0, 1]]
assert one_hot_encode([0, 1, 2]) == [[1, 0, 0],
                                    [0, 1, 0],
                                    [0, 0, 1]]
assert one_hot_encode([2, 2]) == [[0, 0, 1],
                                  [0, 0, 1]]
```
