Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв. Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(text: str) -> int:
    vowels = "aeiouAEIOU"
    count = 0
    for ch in text:
        if ch in vowels:
            count += 1
    return count
# тесты
assert count_vowels("hello") == 2
assert count_vowels("Hi") == 1
assert count_vowels("qwrt") == 0
assert count_vowels("AEIOU") == 5
assert count_vowels("") == 0
```

Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def all_unique(text: str) -> bool:
    for ch in text:
        if text.count(ch) > 1:
            return False
    return True
# тесты
assert all_unique("abc") == True  
assert all_unique("hello") == False  
assert all_unique("") == True        
assert all_unique("Aa") == True      
assert all_unique("test") == False
```

Написать функцию, которая принимает положительное число и возвращает количество бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(n: int) -> int:
    binary = bin(n)  
    return binary.count("1")
# тесты
assert count_bits(1) == 1
assert count_bits(2) == 1      
assert count_bits(3) == 2     
assert count_bits(5) == 2     
assert count_bits(15) == 4     
assert count_bits(1023) == 10  
assert count_bits(0) == 0      

```

Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то, сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы получилось число состоящее из одной цифры. Например, для входного числа: · 39 функция должна вернуть 3, так как 39=27 => 27=14 => 14=4 · 4 функция должна вернуть 0, так как число уже состоит из одной цифры · 999 функция должна вернуть 4, так как 999=729 => 729=126 => 126=12 => 12=2. Написать тесты для кода

```python
def magic(n: int) -> int:
    n = int(n)

    steps = 0
    while n >= 10:
        prod = 1
        for ch in str(n):
            prod *= int(ch)
        n = prod
        steps += 1
    return steps
#тесты
assert magic(39) == 3 
assert magic(999) == 4 
assert magic(4) == 0 

# Доп проверки
assert magic(10) == 1   
assert magic(25) == 2     
assert magic(77) == 4     
assert magic(1) == 0
assert magic(0) == 0      
```

Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Векторы должны быть одинаковой длины")

    s = 0
    for p, t in zip(pred, true):
        s += (p - t) ** 2
    return s / len(pred)
#тесты
assert mse([1, 2, 3], [1, 2, 3]) == 0          
assert mse([1, 2, 3], [2, 2, 2]) == (1 + 0 + 1) / 3

#пример
assert mse([0, 0, 0], [1, 1, 1]) == 1

# Разные числа
assert round(mse([10, 20], [8, 22]), 4) == 4.0

# Отрицательные значения
assert mse([-1, -2], [-1, -3]) == 0.5

# Ошибка длины
try:
    mse([1,2], [1])
except ValueError:
    pass
else:
    raise AssertionError()
```

Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть строку вида “(n1p1)(n2p2)…(nkpk)” представляющая разложение числа на простые множители (если pi == 1, то выводить только ni). Например, для числа 86240 функция должна вернуть “(25)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_decomp(n: int) -> str:

    if n <= 1:
        return ""

    x = n
    p = 2
    parts = []

    while p * p <= x:
        cnt = 0
        while x % p == 0:
            x //= p
            cnt += 1
        if cnt:
            parts.append(f"({p}**{cnt})" if cnt > 1 else f"({p})")
        p += 1 if p == 2 else 2 
        
    if x > 1:
        parts.append(f"({x})")

    return "".join(parts)
# Пример из условия
assert prime_decomp(86240) == "(2**5)(5)(7**2)(11)"

#проверки
assert prime_decomp(1) == ""
assert prime_decomp(2) == "(2)"
assert prime_decomp(3) == "(3)"
assert prime_decomp(4) == "(2**2)"
assert prime_decomp(60) == "(2**2)(3)(5)"
assert prime_decomp(360) == "(2**3)(3**2)(5)"
assert prime_decomp(997) == "(997)"
```

Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    total = 0
    k = 0
    
    while total < number:
        k += 1
        total += k ** 2
    
    if total == number:
        return k
    return "It is impossible"
# Тесты
print(pyramid(1)) 
print(pyramid(5))  
print(pyramid(14)) 
print(pyramid(30)) 
print(pyramid(15))  
print(pyramid(0))   

# Проверки через assert
assert pyramid(1) == 1
assert pyramid(5) == 2
assert pyramid(14) == 3
assert pyramid(30) == 4
assert pyramid(15) == "It is impossible"
assert pyramid(0) == "It is impossible"
```

Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced_number(n):
    s = str(n)
    length = len(s)

    if length <= 2:
        return True

    if length % 2 == 0:  
        mid_left = length // 2 - 1
        mid_right = length // 2
        left = s[:mid_left]
        right = s[mid_right+1:]
    else:
        mid = length // 2
        left = s[:mid]
        right = s[mid+1:]

    return sum(map(int, left)) == sum(map(int, right))

# Примеры
print(is_balanced_number(121)) 
print(is_balanced_number(4518))   
print(is_balanced_number(1238033))
print(is_balanced_number(123))    
print(is_balanced_number(9))      
print(is_balanced_number(22))    


# Проверки через assert
assert is_balanced_number(121) == True
assert is_balanced_number(4518) == True
assert is_balanced_number(1238033) == True
assert is_balanced_number(123) == False
assert is_balanced_number(9) == True
assert is_balanced_number(22) == True
assert is_balanced_number(1234) == False 
```
